```python
#Linear Regression 
sns.regplot(data['CGPA'],data['TOEFL Score'],color ='orange')
plt.show()
```

```python
#Linear Regression 
sns.regplot(data['CGPA'],data['GRE Score'],color ='green')
plt.show()
```

```python
var=['GRE Score','TOEFL Score','University Rating','SOP','LOR ','CGPA','Research']
targ=data['Chance of Admit ']
fig, axes = plt.subplots(7, 1, figsize=(25,40), constrained_layout=True)
fig.suptitle('scatterplot',size=22)
axes=axes.flatten()
for i ,y in enumerate(var):
    ax=axes[i]
    ax.set_xlabel(y , color='blue', size=14)
    ax.set_ylabel('Chances of Admit ', color='red',size=14)
    ax.plot(data[y], targ, marker='o', linestyle='')
```
